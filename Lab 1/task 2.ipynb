# Задача 2 (2.4)

Одеяла с электрообогревом применяются в хирургии для восстановления температуры тела пациента после операции. Имеются два вида одеял: стандартный (b0) и экспериментальный (b1).

Для 14 пациентов известно время, за которое нормальная температура тела восстанавливается при использовании одеяла каждого из видов.

Как понять, отличаются ли экспериментальные одеяла от стандартного?

Требуется:

1. Записать задачу формально в виде проверяемой гипотезы и альтернативы.
2. Предложить не менее 2-х критериев и соответствующих статистик для проверки этой гипотезы и описать:
    * при каких дополнительных условиях (если они есть) стоит применять тот или иной критерий
    * в чём преимущества/недостатки того или иного критерия
3. Аналитически выразить достигаемый уровень значимости каждого критерия на выборке или опишите, как его получить с помощью табличных данных.

Все выкладки должны быть сделаны аналитически, без использования компьютера. (допускается использование компютера для подстановвки численых значений в финальную формулу)

## Решение 

Выборки данных для стандартных и экспериментальных одеял связанные, так как данные собирались с одних и тех же пациентов. Поэтому имеет смысл применять только двухвыборочные критерии для связанных выборок. Будем считать, что одеяла отличаются, если отличаются некоторые средние параметры выборок, например математические ожидания или медианы.

Разберем два случая: (1) характер распределния неизвестен; (2) характер распределния известен - выборки распределены нормально.


### Случай 1

#### 1.1 Постановка задачи

* $X_1^{(n)}, X_2^{(n)}$ - связанные выборки размера n = 14

* Нулевая гипотеза $H_0$: $\mathbb{P}(X_1^{(n)} > X_2^{(n)}) = \frac{1}{2}$

* Альтернативная гипотеза $H_1$: $\mathbb{P}(X_1^{(n)} > X_2^{(n)}) \neq \frac{1}{2}$

#### 1.2 Критерий

Применим двухвыборочный критерий знаков.

* Преимущества: не требует знаний о распределниях, из котороых взяты выборки.

* Недостатки: может проигрывать в мощности параметрическим критериям, как раз из-за того, что не использует информацию о распределении выборок,.

#### 1.3 Статистика

* Cтатистика: $T(X_1^{(n)}, X_2^{(n)}) = \sum\limits_{i=1}^{n} \mathbb{I}(X_{1i} > X_{2i})$

В условиях истинности нулевой гипотезы статистика подчиняется биномиальному закону распределния $Bin(n, p = \frac{1}{2})$.

* $p-value$: $\mathbb{P}(T = k) = \frac{С_n^k}{2^n}$

#### 1.4 Выбор гипотезы 

Пусть задан уровень значимости $\alpha$. Тогда нулевая гипотеза отклоняется тогда и только тогда, когда $p-value \notin [\frac{\alpha}{2}, \frac{1-\alpha}{2}]$.

### Случай 2

#### 2.1 Постановка задачи

* $X_1^{(n)}, X_2^{(n)}$ - связанные выборки из нормального распределния: $X_1^{(n)}$ $\sim$ $N(\mu_1, \sigma_1^2)$, $X_2^{(n)}$ $\sim$ $N(\mu_2, \sigma_2^2)$

* Нулевая гипотеза $H_0$: $\mu_1 = \mu_2$

* Альтернативная гипотеза $H_1$: $\mu_1 \neq \mu_2$

#### 2.2 Критерий

Применим t-критерий Стьюдента для связанных выборок.

* Преимущества: довольно мощный критерий.

* Недостатки: требует знаний о распределнии выборок.

#### 2.3 Статистика

* Cтатистика: $T(X_1^{(n)}, X_2^{(n)}) = \frac{\overline{X}_2 - \overline{X}_2}{S/\sqrt{n}}$, где

$$\overline{X} = \frac{1}{n} \sum\limits_{i=1}^{n} X_i$$ 
$$\overline{D} = \frac{1}{n} \sum\limits_{i=1}^{n} X_{1i} - X_{2i}$$
$$S = \sqrt{\frac{1}{n-1} \sum\limits_{i=1}^{n} (D_i - \overline{D})^2}$$

В условиях истинности нулевой гипотезы статистика подчиняется распределнию Стьюдента $St(n-1)$.

* $p-value$: $min\{\mathbb{P}(T < t), \mathbb{P}(T > t)\}$, где $t = T(X_1^{(n)}, X_2^{(n)})$.

#### 2.4 Выбор гипотезы 

Пусть задан уровень значимости $\alpha$. Тогда нулевая гипотеза отклоняется тогда и только тогда, когда $p-value \notin [\frac{\alpha}{2}, \frac{1-\alpha}{2}]$. С учетом специфики распределния Стьюдента можно переписать это условие в виде:

$$ \mathbb{H}_0 \hspace{5pt} отклоняется \iff t \notin [St(n-1)_{\frac{\alpha}{2}}, St(n-1)_{\frac{1-\alpha}{2}}]$$
